Linear Regression With One Variable</h2>

**Problem Statement**: <p >Building a regression model and predict the per capita income fo canadian citizens in year 2020</p>


year = m * (per capita income) + b


```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df["per capita income (US$)"],color='red',marker='+')
plt.show()
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
per_capita = df["per capita income (US$)"]
per_capita
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita)
```

**(1) Predict per capita income in year = 2006**

```python
reg.predict([[2006]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

**Y = m * X + b (m is coefficient and b is intercept)**
